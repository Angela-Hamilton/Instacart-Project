### Step 3.  Create a regional segmentation fo the data.

```python
# Defining Regions
Northeast = ['Maine', 'New Hampshire', 'Vermont', 'Massachusetts', 'Rhode Island', 'Connecticut', 'New York', 'Pennsylvania', 'New Jersey']
Midwest = ['Wisconsin', 'Michigan', 'Illinois', 'Indiana', 'Ohio', 'North Dakota', 'South Dakota', 'Nebraska', 'Kansas', 'Minnesota', 'Iowa', 'Missouri']
South = ['Delaware', 'Maryland', 'District of Columbia', 'Virginia', 'West Virginia', 'North Carolina', 'South Carolina', 'Georgia', 'Florida', 'Kentucky', 'Tennessee', 'Mississippi', 'Alabama', 'Oklahoma', 'Texas', 'Arkansas', 'Louisiana']
West = ['Idaho', 'Montana', 'Wyoming', 'Nevada', 'Utah', 'Colorado', 'Arizona', 'New Mexico', 'Alaska', 'Washington', 'Oregon', 'California', 'Hawaii']
```

```python
# Assigning states to regions
df_merged_clean.loc[df_merged_clean['state'].isin(Northeast), 'region'] = 'Northeast'
df_merged_clean.loc[df_merged_clean['state'].isin(Midwest), 'region'] = 'Midwest'
df_merged_clean.loc[df_merged_clean['state'].isin(South), 'region'] = 'South'
df_merged_clean.loc[df_merged_clean['state'].isin(West), 'region'] = 'West'
```

```python
# Checking frequency
df_merged_clean['region'].value_counts(dropna = False)
```

```python
df_merged_clean.head()
```

### Check for variations in spending habits by region

```python
# Crossing variables
crosstab = pd.crosstab(df_merged_clean['region'], df_merged_clean['spending_flag'], dropna = False)
```

```python
crosstab
```

```python
# Copy to clipboard
crosstab.to_clipboard()
```

#### The South Region appears to spend more than other regions with a spending habit of 33%, whereas the Northeast Region is considered the lowest spender with a spending habit of approximately 18% which is lower than all other regions.

### Step 4. Create an exclusion flag for low-activity customers with less than 5 orders and exclude them from the data.

```python
# Create flag for customers with less than 5 orders
df_merged_clean.loc[df_merged_clean['max_order'] < 5, 'customer_activity'] = 'low activity'
```

```python
# Create flag for customers with >=5 orders
df_merged_clean.loc[df_merged_clean['max_order'] >= 5, 'customer_activity'] = 'high activity'
```

```python
# Check frequency
df_merged_clean['customer_activity'].value_counts(dropna = False)
```

```python
# Assign low activity profiles to new dataframe
low_activity_customers = df_merged_clean[df_merged_clean['customer_activity'] == 'low activity']
```

```python
# Export and save low activity dataframe
low_activity_customers.to_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'low_activity_customers.pkl'))
```

```python
# Assign high activity profiles to new dataframe
high_activity_customers = df_merged_clean[df_merged_clean['customer_activity'] == 'high activity']
```

```python
# Export and save high activity profiles
high_activity_customers.to_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'high_activity_customers.pkl'))
```
